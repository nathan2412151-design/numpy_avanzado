```python
import numpy as np
```

```python
# ejercicio 1
import numpy as np

a = np.array([1, 2, 3, 4])

resultado = np.all(a != 0)
print(resultado)
```

```python
a = np.array([0, 0, 2, 0])
resultado = np.any(a != 0)
print(resultado)
```

```python
a = np.array([1, 2, np.nan, np.inf])

tiene_nan = np.any(np.isnan(a))
tiene_inf = np.any(np.isinf(a))

print("¿Tiene NaN?", tiene_nan)
print("¿Tiene infinitos?", tiene_inf)
```

```python
a = np.array([1, 2, np.inf])

tiene_inf = np.any(np.isinf(a))

print(tiene_inf)
```

```python
# ejercicio 5
a = np.array([1, np.nan, 3])

tiene_nan = np.any(np.isnan(a))
print(tiene_nan)
```

```python
# ejercicio 6
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

print("a > b:", a > b)
print("a >= b:", a >= b)
print("a < b:", a < b)
print("a <= b:", a <= b)
```

```python
# ejercicio 7
a = np.array([1, 7, 13, 105])
print(a.nbytes)
```

```python
# ejercicio 8
ceros = np.zeros(10)
unos = np.ones(10)
cincos = np.full(10, 5)

print(ceros)
print(unos)
print(cincos)
```
